```python
import json
import pandas as pd

# 
train_jsonl_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/MR_train_queries.jsonl'
valid_jsonl_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/MR_valid_queries.jsonl'
new_train_jsonl_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/train_queries.jsonl'
new_valid_jsonl_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/valid_queries.jsonl'

train_tsv_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/MR_train_imgs.tsv'
valid_tsv_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/MR_valid_imgs.tsv'
new_train_tsv_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/train_imgs.tsv'
new_valid_tsv_path = '/mnt/petrelfs/lijiaxing1/lijiaxing/Chinese-CLIP/data/datasets/Li_data/valid_imgs.tsv'

#  JSONL 
def read_jsonl(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    return [json.loads(line.strip()) for line in lines]

#  JSONL 
def write_jsonl(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as f:
        for entry in data:
            f.write(json.dumps(entry, ensure_ascii=False) + '\n')

#  TSV 
def read_tsv(file_path):
    return pd.read_csv(file_path, sep='\t', header=None)

#  TSV 
def write_tsv(file_path, df):
    df.to_csv(file_path, sep='\t', header=False, index=False)

# 
train_data = read_jsonl(train_jsonl_path)
valid_data = read_jsonl(valid_jsonl_path)
train_tsv = read_tsv(train_tsv_path)
valid_tsv = read_tsv(valid_tsv_path)

#  query_id 
query_id_counter = max([entry['query_id'] for entry in train_data] + [entry['query_id'] for entry in valid_data]) + 1

# 
new_train_data = train_data.copy()
new_valid_data = valid_data.copy()

#  4000 
for i in range(4000):
    valid_entry = new_valid_data[i]
    query_text = valid_entry['query_text']
    item_ids = valid_entry['item_ids']
    
    #  entry  item_ids
    items_to_add = item_ids[:2]
    remaining_items = item_ids[2:]

    #  entry
    new_valid_data[i]['item_ids'] = remaining_items

    #  item_ids 
    for item_id in items_to_add:
        new_train_entry = {
            "query_id": query_id_counter,
            "query_text": query_text,
            "item_ids": [item_id]
        }
        new_train_data.append(new_train_entry)
        query_id_counter += 1

    #  TSV  item_ids  TSV 
    for item_id in items_to_add:
        item_row = valid_tsv[valid_tsv[0] == item_id]
        valid_tsv = valid_tsv[valid_tsv[0] != item_id]
        train_tsv = pd.concat([train_tsv, item_row])

#  JSONL  TSV 
write_jsonl(new_train_jsonl_path, new_train_data)
write_jsonl(new_valid_jsonl_path, new_valid_data)
write_tsv(new_train_tsv_path, train_tsv)
write_tsv(new_valid_tsv_path, valid_tsv)

print("")
```
